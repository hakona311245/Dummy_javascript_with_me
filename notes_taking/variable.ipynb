### Variable in js

Alright so it's weird especially coming from java, C#, C++ or even python

But in js when you need a variable you just declare let (or older way is var) and you can use it for any data types

```python
let user = 'John';
let age = 25;
let message = 'Hello';
```

Then you can do operation quite relaxing:

```python
let hello = 'Hello world!';

let message;

// copy 'Hello world' from hello into message
message = hello;

// now two variables hold the same data
alert(hello); // Hello world!
alert(message); // Hello world!
```

There are two limitations on variable names in JavaScript:

The name must contain only letters, digits, or the symbols $ and _.
The first character must not be a digit.

```python
let $ = 1; // declared a variable with the name "$"
let _ = 2; // and now a variable with the name "_"

alert($ + _); // 3
```

```python
Also js is case-sensitive. So, for instance, the variables apple and Apple are different.
```
